# 2. Credenciales PAT en Colab


Si no tenes Personal Access Token ir al siguiente [link](https://github.com/settings/tokens), generar un token con permisos de acceso. Y copiar el token al Colab.

Aca presento dos opciones, una es tener el token guardado en un txt para futuros usos y otra es insertando el token por medio de la lib getpass.

```python
repo_to_clone = "shared-resources"
```

```python
# Opcion 1 : Tener un TXT con el TOKEN en DRIVE
read_token_from_file = True
if read_token_from_file:
  filename = 'token_github_personal.txt'
  path_to_token = f"{repo_drive_path}/{filename}"

  with open(path_to_token, 'r') as file:
      token = file.read().strip()
else:
  from getpass import getpass
  token = getpass('Introduce tu token de acceso: ')


!git clone https://{token}@github.com/{github_username}/{repo_to_clone}.git
```
